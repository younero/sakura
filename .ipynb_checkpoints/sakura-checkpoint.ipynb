```python
# 读取源视频，逐帧截取为图片以便之后的处理
import cv2
vidcap = cv2.VideoCapture('video/video.avi')
success,image = vidcap.read()
count = 0
success = True
while success:
    success,image = vidcap.read()
    cv2.imwrite("pic/frame%d.jpg" % count, image)  # save frame as JPEG file
    if cv2.waitKey(10) == 27:
        break
    count += 1
```

```python
from PIL import Image, ImageDraw, ImageFont
import cv2
import os


def draw(pic):
    img = cv2.imread('pic/' + pic)
    img = img[:, :, (2, 1, 0)]

    blank = Image.new("RGB", [len(img[0]), len(img)], "white")
    drawObj = ImageDraw.Draw(blank)

    n = 10

    font = ImageFont.truetype('font/msyhbd.ttc', size=n - 1)

    for i in range(0, len(img), n): # 间隔n=10个像素块读取颜色（源图片分辨率太高了最后生成的文字会有重叠）
        for j in range(0, len(img[i]), n):
            text = '武汉加油'
            drawObj.ink = img[i][j][0] + img[i][j][1] * 256 + img[i][j][2] * 256 * 256 # 根据图片对应位置的颜色上色
            drawObj.text([j, i], text[int(j / n) % len(text)], font=font) # 按“武汉加油”的顺序填充文字

    blank.save('new/new_' + pic, 'jpeg') # 保存生成的图片


filelist = os.listdir('pic')
n = 1
for file in filelist:
    try:
        draw(file)
        print('finish %d/%d'%(n, len(filelist)))
    except:
        print('error %d'%n)
    n+=1
```

```python
import time
import os
import cv2
import re


def resort(list):
    for i in range(len(list)-1):
        for j in range(len(list)-1):
            if int(re.findall(r'\d+', list[j])[0]) > int(re.findall(r'\d+', list[j+1])[0]):
                list[j], list[j+1] = list[j+1], list[j]
    return list


def picvideo(path, size):
    filelist = os.listdir(path)
    filelist = resort(filelist) # 读取到的图片列表可能不是按顺序的，需要重新排一下序

    
    fps = 24 # 视频帧率

    file_path = 'video/new.mp4'
    fourcc = cv2.VideoWriter_fourcc('D', 'I', 'V', 'X')  # MP4的编码格式

    video = cv2.VideoWriter(file_path, fourcc, fps, size) # size是图片的分辨率

    for item in filelist:
        if item.endswith('.jpg'):
            item = path + '/' + item
            img = cv2.imread(item)
            video.write(img)

    video.release()
    print('done!')


picvideo(r'new', (960, 544))
```
